```python
import pandas as pd
import glob
```

```python
files = glob.glob("outputs/uncertainty**.pcl")
files
```

```python
df = pd.DataFrame()
for f in files:
    new_data = pd.read_pickle(f)
    new_data['Name'] = f.split("uncertainty_")[1].split('.pcl')[0]
    if "largearea" in new_data['Name'].iloc[0]:
        new_data['Name'] = new_data['Name'].iloc[0].split("largearea_")[1] + " (large area)"
    if "wholemountain" == new_data['Name'].iloc[0]:
        new_data['Name'] = "whole mountain"
    new_data['Name'] = new_data['Name'].apply(lambda x: x.title())
    df = df.append(new_data)
```

```python
pd.set_option('display.max_rows', 100)
```

```python
df = df[['Name', 'Start Date', 'End Date', 'Mean', 'RMSE', 'Range', 'Sill', '90% CI Lower Bound', '90% CI Upper Bound']].round(decimals=2)
```

```python
src = df[df['Name'].str.contains("Large Area")]
src = src[src['Start Date'].dt.year==1947]
src
```

```python
df.reset_index(drop=True)
```

```python
df.to_csv("outputs/uncertainty.csv", index=False)

```
